```python
# **Please Complete this Coding Exercise:-**

#1. Concatenate these 3 arrays into a new array named 'newArray'...
print("\n#1. - Concatenate arrays")
import numpy as np
import pandas as pd

Array1 = np.array([[25, 16]])
Array2 = np.array([[11, 2], [13, 4]])
Array3 = np.array([[7, 81], [5, 6], [11, 12]])

newArray = np.concatenate([Array1, Array2, Array3])
print(newArray)


#2. Sort 'newArray' in order into 'sortedArray'
print("\n#2. - Sort array")
print("\n Sorted array")
sortedArray = np.sort(newArray)
print(sortedArray)

#3. Reshape the 'sortedArray' array, into a new array called 'reshapedArray', 
# so it has 2 dimensions with a size of 2, 3.
print("\n#3. - Reshape sorted array")

reshapedArray = np.reshape(sortedArray, (2,2,3))
print( "\n 2-D reshaped 2x3 array\n", reshapedArray)

#4. Unpack the array tuples from the above 'reshapedArray'  
#   into 4 well named variables. Print the 4 variables.
print("\n#4. - Unpack the array")

a, b = reshapedArray
a1, a2 = a
b1, b2 = b
print("\n Unpacked tuples from previous array:")
print("a1 = ", a1)
print("a2 = ", a2)
print("b1 = ", b1)
print("b2 = ", b2)

```
